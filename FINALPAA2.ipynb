```python
# BUBBLE SORT DAN INSERTION SORT
import time

def bubble_sort(arr):
    n = len(arr)
    start_time = time.time()
    for i in range(n):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
        if i < 5 or i >= n - 5:
            print(f"Iterasi {i + 1}: {arr}")
    end_time = time.time()
    duration = end_time - start_time
    return duration

def insertion_sort(arr):
    n = len(arr)
    start_time = time.time()
    for i in range(1, n):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
        if i < 5 or i >= n - 5:
            print(f"Iterasi {i + 1}: {arr}")
    end_time = time.time()
    duration = end_time - start_time
    return duration

def main():
    arr = [12, 99, 62, 15, 20, 95, 39, 48, 3, 24, 8, 43, 74, 19, 97, 33, 49, 68, 55, 33, 90, 90, 7, 26, 85, 46, 39, 40, 9, 36, 60, 64, 89, 31, 25, 71, 21, 23, 63, 84, 32, 5, 3, 44, 21, 10, 21, 17, 50, 2, 36, 53, 79, 54, 19, 88, 1, 32, 31, 15, 6, 3, 1, 40, 22, 43, 18, 1, 77, 9, 59, 40, 7, 41, 81]

    print("Array sebelum pengurutan:", arr)

    choice = ''
    while choice != '3':
        print("\nMenu Pilihan:")
        print("1. Bubble Sort")
        print("2. Insertion Sort")
        print("3. Keluar")

        choice = input("Pilih jenis pengurutan (1-3): ")

        if choice == '1':
            bubble_sort_duration = bubble_sort(arr.copy())
            sort_type = "Bubble Sort"
            print(f"\nWaktu komputasi Bubble Sort: {bubble_sort_duration:.6f} detik")
            print("Array setelah pengurutan:", arr)
        elif choice == '2':
            insertion_sort_duration = insertion_sort(arr.copy())
            sort_type = "Insertion Sort"
            print(f"\nWaktu komputasi Insertion Sort: {insertion_sort_duration:.6f} detik")
            print("Array setelah pengurutan:", arr)
        elif choice == '3':
            print("Keluar")
        else:
            print("Pilihan tidak valid. Silakan coba lagi.")

if __name__ == "__main__":
    main()
```

```python
# TSP DAN DJIKSTRA
import time
import itertools
from heapq import heappop, heappush

def tsp(graph):
    start_time = time.time()

    nodes = list(graph.keys())
    permutations = list(itertools.permutations(nodes))

    shortest_path = None
    shortest_distance = float('inf')

    iteration_count = 0

    for path in permutations:
        distance = 0
        valid_path = True
        for i in range(len(path) - 1):
            current_node = path[i]
            next_node = path[i + 1]

            if next_node not in graph[current_node]:
                valid_path = False
                break

            distance += graph[current_node][next_node]

        if valid_path and distance < shortest_distance:
            shortest_distance = distance
            shortest_path = path

        iteration_count += 1
        print(f"Iterasi ke-{iteration_count}: {path} - Jarak: {distance}")

    end_time = time.time()
    computation_time = end_time - start_time

    print("\nHasil Akhir TSP:")
    print(f"Jalur Terpendek: {shortest_path}")
    print(f"Jarak Terpendek: {shortest_distance}")
    print(f"Waktu Komputasi: {computation_time} detik")

def dijkstra(graph):
    start_time = time.time()

    distances = {node: {n: float('inf') for n in graph} for node in graph}

    for node in graph:
        distances[node][node] = 0
        heap = [(0, node)]

        while heap:
            current_distance, current_node = heappop(heap)

            if current_distance > distances[node][current_node]:
                continue

            for neighbor, weight in graph[current_node].items():
                distance = current_distance + weight

                if distance < distances[node][neighbor]:
                    distances[node][neighbor] = distance
                    heappush(heap, (distance, neighbor))

    end_time = time.time()
    computation_time = end_time - start_time

    print("\nHasil Akhir Dijkstra:")
    for node in distances:
        print(f"Jarak Terpendek dari node {node}:")
        for n in distances[node]:
            print(f"{node} ke {n}: {distances[node][n]}")
        print()

    print(f"Waktu Komputasi: {computation_time} detik")

def main():
    graph = {
        'a': {'b': 12, 'c': 10, 'g': 12},
        'b': {'a': 12, 'c': 8, 'd': 12},
        'c': {'a': 10, 'b': 8, 'd': 11, 'e': 3, 'g': 9},
        'd': {'b': 12, 'c': 11, 'e': 11, 'f': 10},
        'e': {'c': 3, 'd': 11, 'f': 6, 'g': 7},
        'f': {'d': 10, 'e': 6, 'g': 9},
        'g': {'a': 12, 'c': 9, 'e': 7, 'f': 9}
    }

    while True:
        print("\nPilih algoritma:")
        print("1. TSP")
        print("2. Dijkstra")
        print("3. Keluar")
        user_choice = input("Masukkan pilihan: ")

        if user_choice == "1":
            tsp(graph)
        elif user_choice == "2":
            dijkstra(graph)
        elif user_choice == "3":
            break
        else:
            print("Pilihan tidak valid. Silakan pilih 1, 2, atau 3.")

if __name__ == "__main__":
    main()
```
